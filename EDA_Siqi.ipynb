```python
plt.scatter(x=data_df['age'], y=data_df['charges'], alpha=0.5)


# Decorate
plt.title('Age Of person')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
sns.countplot(data_df['sex'])
```

```python
sns.countplot(data_df['region'])
```

```python
plt.scatter(x=data_df['bmi'], y=data_df['charges'], alpha=0.5)


# Decorate
plt.title('Bmi Of person')
plt.xlabel('Bmi')
plt.ylabel('Charges')
plt.show()
```

```python
plt.scatter(x=data_df['smoker'], y=data_df['charges'], alpha=0.5)


# Decorate
plt.title('Smoker vs. Charges')
plt.xlabel('Smokers')
plt.ylabel('Charges')
plt.show()
```

```python
plt.scatter(x=data_df['children'], y=data_df['charges'], alpha=0.5)


# Decorate
plt.title('Children vs. Charges')
plt.xlabel('Children')
plt.ylabel('Charges')
plt.show()
```

```python
sns.set(style='whitegrid')
f, ax = plt.subplots(1,1, figsize=(12, 8))
ax = sns.distplot(data_df['charges'], kde = True, color = 'c')
plt.title('Distribution of Charges')
```

```python
ax = sns.lmplot(x = 'age', y = 'charges', data=data_df, hue='smoker', palette='Set1')
ax = sns.lmplot(x = 'bmi', y = 'charges', data=data_df, hue='smoker', palette='Set2')
ax = sns.lmplot(x = 'children', y = 'charges', data=data_df, hue='smoker', palette='Set3')
```

```python
# Create a heatmap
f, ax = plt.subplots(1, 1, figsize=(10, 10))
ax = sns.heatmap(data_df.corr(), annot=True, cmap='Blues')
```
