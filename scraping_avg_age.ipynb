```python
# Import libraries for scraping 
from splinter import Browser
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd
```

```python
# Selenium v4 compatible Code Block
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.google.com")
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit website to extract data (https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_life_expectancy)
url = 'https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_life_expectancy'
page = browser.visit(url)
# Optional delay for loading the page
browser.is_element_present_by_css('div.list_text', wait_time=1)
```

```python
html = browser.html

news_soup = soup(html, 'html.parser')
slide_elem = news_soup.select_one('div.list_text')
```

```python
life_expectancy_df = pd.read_html('https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_life_expectancy')[0]
life_expectancy_df.columns=['rank','state', 'life expectency']
life_expectancy_df.set_index('state', inplace=False)

life_expectancy_df = life_expectancy_df.iloc[1: , :]

life_expectancy_df.head(5)
```

```python
browser.quit()
```

```python
#removing "rank" column
life_expectancy_df = life_expectancy_df.drop("rank", axis=1)
```

```python
# removing counties and cleaning data from table 

life_expectancy_df.drop(life_expectancy_df[(life_expectancy_df['state'] == 'Guam[2]') | (life_expectancy_df['state'] == 'Puerto Rico[2]') | (life_expectancy_df['state'] == 'U.S. Virgin Islands[2]') | (life_expectancy_df['state'] == 'United States') | (life_expectancy_df['state'] == 'Northern Mariana Islands[6]')| (life_expectancy_df['state'] == 'American Samoa[6]')].index, inplace= True)
```
