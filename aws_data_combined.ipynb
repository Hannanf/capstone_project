```python
# Import our dependencies
import pandas as pd
import sqlalchemy as sql
```

```python
# Connecting to PostgreSQL database.
engine = sql.create_engine('postgresql://postgres:password@localhost:5432/insurance')
```

```python
# Getting the table names stored.
engine.table_names()
```

```python
# Reading data stored localhost.
complete_data_df = pd.read_sql_table('combined_dataframes', con=engine)
complete_data_df
```

```python
complete_data_df['average_income'] = complete_data_df['average_income'].str.replace(",", "")
complete_data_df['average_income'] = complete_data_df['average_income'].str.replace("$", "")
complete_data_df['average_income'] = complete_data_df['average_income'].astype(float)
```

```python
complete_data_df.nunique()
```

```python
complete_data_cat = complete_data_df.dtypes[complete_data_df.dtypes == "object"].index.tolist()
complete_data_df
```

```python
# Getting the table names stored.
engine1.table_names()
```

```python
# Reading data stored localhost.
average_age_by_state_df = pd.read_sql_table('average_age_by_state', con=engine1)
average_age_by_state_df
```
